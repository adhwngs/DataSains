```python
df_gptest = df[['BMI Category','Quality of Sleep','Stress Level']]
grouped_test1 = df_gptest.groupby(['BMI Category','Quality of Sleep'],as_index=False).mean()
grouped_test1
```

```python
grouped_pivot = grouped_test1.pivot(index='BMI Category',columns='Quality of Sleep')
grouped_pivot
```

```python
grouped_pivot = grouped_pivot.fillna(0) #fill missing values with 0
grouped_pivot
```

```python
df_gptest2 = df[['Quality of Sleep','Stress Level']]
grouped_test_bodystyle = df_gptest2.groupby(['Quality of Sleep'],as_index= False).mean()
grouped_test_bodystyle
```

```python
plt.pcolor(grouped_pivot, cmap='RdBu')
plt.colorbar()
plt.show()
```

```python
fig, ax = plt.subplots()
im = ax.pcolor(grouped_pivot, cmap='RdBu')

# nama label
row_labels = grouped_pivot.columns.levels[1]
col_labels = grouped_pivot.index

# pindahkan ticks and labels ke tenah
ax.set_xticks(np.arange(grouped_pivot.shape[1]) + 0.5, minor=False)
ax.set_yticks(np.arange(grouped_pivot.shape[0]) + 0.5, minor=False)

# masukan  labels
ax.set_xticklabels(row_labels, minor=False)
ax.set_yticklabels(col_labels, minor=False)

# rotasi label
plt.xticks(rotation=90)

fig.colorbar(im)
plt.show()
```

**ANOVA**

```python
grouped_test2=df_gptest[['BMI Category','Quality of Sleep']].groupby(['BMI Category'])
grouped_test2.head(2)
```

```python
df_gptest
```

```python
from scipy import stats

f_val, p_val = stats.f_oneway(
    grouped_test2.get_group('Normal')['Quality of Sleep'],
    grouped_test2.get_group('Normal Weight')['Quality of Sleep'],
    grouped_test2.get_group('Overweight')['Quality of Sleep'],
    grouped_test2.get_group('Obese')['Quality of Sleep']
)

print("ANOVA results: F =", f_val, ", P =", p_val)
```
